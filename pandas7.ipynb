```python
import pandas as pd
import numpy as np

nparray  = np.random.randint(1,15,size=(5,3))
nparray2  = np.random.randint(1,15,size=(5,3))

df1=pd.DataFrame(nparray,columns=["s1","s2","s3"])
df2=pd.DataFrame(nparray2,columns=["s1","s2","s3"])
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
df2["s4"]=[0,1,2,3,4]
pd.concat([df1,df2],ignore_index=True)
```

```python
pd.concat([df1,df2],ignore_index=True,join="inner")
```
